### 範例一

繪出微分方程式
$$
\frac{dy}{dx} = \frac{2x+1}{2y}
$$
的方向場，以及通過初始條件 $y(-2)=-1$ 的解曲線。

#### 解答

繪出方向場：

```python
# 定義變數
x, y = var('x, y')
Plot1 = plot_slope_field((2*x+1)/(2*y), (x, -5, 5), (y, -5, 5), aspect_ratio=1)
Plot1.show()
```

解出初始值問題：

```python
%display latex

# 定義函數關係
y = function('y')(x)

# 定義微分方程式
DE1 = diff(y,x) == (2*x+1)/(2*y)
Sol1 = desolve(DE1, y, ics=[-2,-1])
Sol1.show()
```

由於解是隱式 (implicit form)，故需要使用 `solve` 解出 $y = y(x)$:

```python
# 使用 solve 解出 y(x)
Sol1_2 = solve(Sol1,y)
Sol1_2
```

由於初始條件為 $y(-2)=-1$，代表解位於第三象限 $(y \le 0)$。取出第一個解之等號右側 (rhs, right-hand side) 來繪圖：

```python
# Sol1_2[0] 代表其中的第一個元素（第一個解）；.rhs() 代表取出等號右側。注意繪圖範圍：
Plot1_2 =  plot(Sol1_2[0].rhs(), x, -5, 0)
Plot1_2.show()
```

結合方向場與解曲線：

```python
Plot1 = Plot1 + Plot1_2
Plot1.show()
```

### 範例二

$$
x^2 \frac{dy}{dx} = y - xy,\quad y(-1) = -1
$$

#### 解答

繪出分向場：

```python
x, y = var('x, y')
Plot1 = plot_slope_field((y-x*y)/(x^2), (x,-5,5), (y,-5,5), aspect_ratio=1)
Plot1.show()
```

解出 IVP:

```python
y = function('y')(x)
DE = diff(y,x) == (y-x*y)/(x^2)
Sol = desolve(DE, y, ics=[-1,-1])
Sol
```

由於是顯式解，故可以直接使用結果來畫圖。但要注意 x 的範圍：

```python
Plot2 = plot(Sol, x, -5, -0.5)
Plot2.show()
```

```python
Plot1 = Plot1 + Plot2
Plot1.show()
```
