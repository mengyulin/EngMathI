ä¸€å€‹ç¬¦è™Ÿé‹ç®—å¼å¯ä»¥åƒåœ– 2.1 é‚£æ¨£è¢«çœ‹ä½œæ˜¯ä¸€æ£µæ¨¹ç‹€çµæ§‹ä¾†è¡¨é”ã€‚

é‡è¦çš„æ˜¯è¦ç†è§£ï¼šç¬¦è™Ÿé‹ç®—å¼æ˜¯ä¸€å€‹æ–¹ç¨‹å¼ï¼Œè€Œä¸æ˜¯ä¸€å€‹å…·é«”çš„å€¼æˆ–æ•¸å­¸å‡½æ•¸ã€‚å› æ­¤ï¼ŒSage ä¸¦ä¸æœƒå°‡ä»¥ä¸‹å…©å€‹é‹ç®—å¼è­˜åˆ¥ç‚ºç›¸ç­‰ã€‚

>* ç›¸ç­‰æ€§æ¸¬è©¦ == ä¸åƒ…åƒ…æ˜¯èªæ³•ä¸Šçš„æ¯”è¼ƒã€‚ä¾‹å¦‚ï¼šé‹ç®—å¼ arctan(sqrt(2)) å’Œ pi/2 - arctan(1/sqrt(2)) è¢«è¦–ç‚ºç›¸ç­‰ã€‚ä½†å¯¦éš›ä¸Šï¼Œä½¿ç”¨ bool(x == y) æ¯”è¼ƒå…©å€‹é‹ç®—å¼ï¼ŒSage æœƒå˜—è©¦è­‰æ˜å®ƒå€‘çš„æ˜¯å¦å®Œå…¨ç›¸ç­‰ï¼Œå¦‚æœæˆåŠŸè­‰æ˜ç‚ºé›¶è¡¨ç¤ºå·®ç•°ç‚ºé›¶ï¼Œå‰‡è¿”å› Trueã€‚ä¸å®Œå…¨ç›¸ç­‰å‰‡è¿”å› Falseã€‚

æœ€å¸¸è¦‹çš„ä½¿ç”¨æ–¹å¼ï¼Œæ˜¯é€šéå®šç¾©é‹ç®—å¼ä¸­çš„æŸäº›åƒæ•¸ä¾†é€²è¡Œæ±‚å€¼ã€‚æ–¹æ³• `subs` å¯ä»¥åŸ·è¡Œé€™ç¨®è½‰æ›ï¼Œè€Œä¸”é€™å€‹æ–¹æ³•å¯ä»¥ç”¨ä¾†è¡¨é”éš±å‡½æ•¸ï¼š

```python
# å®£å‘Š a, x ç‚ºè®Šæ•¸
a, x = var('a, x')

# å°‡ä¸‹è¿°ç¬¦è™Ÿé‹ç®—å¼å‘½åç‚º y
y = cos(x+a) * (x+1)

# è¼¸å‡ºæ­¤ç¬¦è™Ÿé‹ç®—å¼
y
```

ç›¸æ¯”æ–¼å¸¸è¦‹çš„æ•¸å­¸ç¬¦è™Ÿè¡¨ç¤º ğ‘¥ â†¦ ğ‘“(ğ‘¥)ï¼Œåœ¨ Sage ä¸­ï¼Œè¢«æ›¿æ›çš„è®Šæ•¸å¿…é ˆæ˜ç¢ºå®šç¾©ã€‚å¤šå€‹è®Šæ•¸çš„æ›¿æ›æ˜¯ä¸¦è¡Œé€²è¡Œçš„ï¼Œè€Œé€£çºŒçš„æ›¿æ›å‰‡æ˜¯ä¾æ¬¡åŸ·è¡Œçš„ï¼Œå¦‚ä¸‹é¢çš„å…©å€‹ä¾‹å­æ‰€ç¤ºï¼š

```python
x, y, z = var('x, y, z') ; q = x*y + y*z + z*x

# å¤šå€‹è®Šæ•¸çš„æ›¿æ›
bool(q(x=y, y=z, z=x) == q)
```

### 2.1.2 é‹ç®—å¼çš„è½‰æ›
æœ€ç°¡å–®çš„éå¸¸æ•¸é‹ç®—å¼ï¼Œæ˜¯åŒ…å«ä¸€å€‹æˆ–å¤šå€‹è®Šæ•¸çš„å¤šé …å¼å’Œæœ‰ç†å‡½æ•¸ã€‚å°‡é‹ç®—å¼é‡å¯«ç‚ºå¤šç¨®å½¢å¼ã€æˆ–å°‡å…¶åŒ–ç‚ºæ¨™æº–å½¢å¼çš„å‡½æ•¸ç¸½çµæ–¼è¡¨ 2.1 ä¸­ã€‚ä¾‹å¦‚ï¼šæ–¹æ³•`expand` éå¸¸é©åˆç”¨æ–¼å±•é–‹å¤šé …å¼ï¼š

```python
# è¨­å®š x, y ç‚ºç¬¦è™Ÿè®Šæ•¸ã€‚æ³¨æ„ï¼švar() èˆ‡ SR.var() æ„ç¾©ç›¸åŒã€‚ 
# (SR æ„æŒ‡ Symbolic Ring, è€Œ ring ç‚ºä»£æ•¸ç†è«–çš„ç”¨è©)
x, y = SR.var('x,y')
```

```python
# å®šç¾©ç¬¦è™Ÿé‹ç®—å¼ä¸¦å‘½åç‚º p
p = (x+y)*(x+1)^2

# å®šç¾© p2 ç‚º p çš„å±•é–‹å½¢å¼
p2 = p.expand(); p2
```

è€Œæ–¹æ³• `collect` å‰‡æœƒå°‡å„é …ä¾ç…§åƒæ•¸çš„å†ªæ¬¡é€²è¡Œåˆ†çµ„æ’åˆ—ã€‚

* è¨»ï¼š`method(parameter)`æ–¹æ³•å¾Œæ–¹æ‹¬è™Ÿä¸­çš„è®Šæ•¸ç¨±ç‚ºæ­¤æ–¹æ³•çš„åƒæ•¸ã€‚

```python
# ä¾ç…§ x çš„å†ªæ¬¡æ’åˆ—
p2.collect(x)
```

é€™äº›æ–¹æ³•ä¸åƒ…é©ç”¨æ–¼ç¬¦è™Ÿè®Šæ•¸ä¸­çš„å¤šé …å¼ï¼Œé‚„é©ç”¨æ–¼æ›´è¤‡é›œçš„å­é‹ç®—å¼ï¼Œä¾‹å¦‚ `sin(x)`ï¼š

```python
# å°‡é‹ç®—å¼ä½¿ç”¨æ–¹æ³• `expand` å±•é–‹ï¼Œå†ç”¨æ–¹æ³• `collect` ä¾ç…§åƒæ•¸ `sin(x)` çš„å†ªæ¬¡åˆ†çµ„æ’åˆ—ã€‚
((x+y+sin(x))^2).expand().collect(sin(x))
```

ä¸‹è¡¨æ•´ç†äº†å¤šé …å¼èˆ‡åˆ†å¼çš„å±•é–‹èˆ‡è½‰æ›æ–¹æ³•ï¼š



å°æ–¼æœ‰ç†å‡½æ•¸ï¼Œæ–¹æ³• `combine` å¯ä»¥å°‡å…·æœ‰ç›¸åŒåˆ†æ¯çš„é …ç›®åˆä½µï¼›æ–¹æ³• `partial_fraction` å‰‡å¯ä»¥åœ¨ ğ‘„ ä¸Šé€²è¡Œéƒ¨åˆ†åˆ†å¼åˆ†è§£ã€‚
å°å¤šé …å¼ä¾†èªªï¼Œæ›´æœ‰ç”¨çš„è¡¨ç¤ºå½¢å¼æ˜¯å±•é–‹å½¢å¼ï¼›è€Œå°æ–¼åˆ†å¼ï¼Œå‰‡æ˜¯å°‡ ğ‘ƒ/ğ‘„ ä¸­çš„ ğ‘ƒ å’Œ ğ‘„ å±•é–‹å¾Œçš„ç°¡åŒ–å½¢å¼ã€‚ç•¶å…©å€‹å¤šé …å¼æˆ–åˆ†å¼ä»¥é€™ç¨®å½¢å¼è¡¨ç¤ºæ™‚ï¼Œåªéœ€æ¯”è¼ƒå®ƒå€‘çš„ä¿‚æ•¸å³å¯åˆ¤æ–·å®ƒå€‘æ˜¯å¦ç›¸ç­‰ï¼šæˆ‘å€‘ç¨±é€™ç¨®å½¢å¼ç‚ºæ¨™æº–å½¢å¼ï¼ˆnormal formï¼‰ã€‚

#### ç¬¦è™Ÿå‡½æ•¸
Sage ä¹Ÿå…è¨±å®šç¾©ç¬¦è™Ÿå‡½æ•¸ä¾†æ“ä½œé‹ç®—å¼ï¼š

```python
x = var('x')

# å®šç¾©ç¬¦è™Ÿå‡½æ•¸ f(x) ä¸¦è¼¸å‡º f(-3)
f(x)=(2*x+1)^3 ; f(-3)
```

```python
# å°‡ f(x) å±•é–‹
f.expand()
```

ç¬¦è™Ÿå‡½æ•¸å°±åƒæ˜¯ä¸€å€‹å‘¼å«æŒ‡ä»¤çš„é‹ç®—å¼ï¼Œå…¶ä¸­åƒæ•¸çš„é †åºæ˜¯å›ºå®šçš„ã€‚è¦å°‡ç¬¦è™Ÿé‹ç®—å¼è½‰æ›ç‚ºç¬¦è™Ÿå‡½æ•¸ï¼Œå¯ä»¥ä½¿ç”¨ `f(x) = ...` æˆ–æ–¹æ³• `function` ï¼š

```python
x, y = var('x, y')

# å®šç¾© u ç‚ºç¬¦è™Ÿé‹ç®—å¼
u = sin(x) + x*cos(y)

# å°‡ u ä»£è¡¨çš„é‹ç®—å¼è½‰æ›ç‚ºç¬¦è™Ÿå‡½æ•¸ä¸¦å‘½åç‚º v
v = u.function(x, y); v
```

### 2.1.3 å¸¸è¦‹çš„æ•¸å­¸å‡½æ•¸
Sage æ”¯æ´å¤§å¤šæ•¸çš„æ•¸å­¸å‡½æ•¸ï¼Œç‰¹åˆ¥æ˜¯ä¸‰è§’å‡½æ•¸ã€å°æ•¸å’ŒæŒ‡æ•¸å‡½æ•¸ç­‰ï¼Œé€™äº›å‡½æ•¸ç¸½çµåœ¨è¡¨ 2.2 ä¸­ã€‚

äº†è§£å¦‚ä½•è½‰æ›é€™äº›å‡½æ•¸æ˜¯éå¸¸é‡è¦çš„ã€‚è¦ç°¡åŒ–é‹ç®—å¼æˆ–ç¬¦è™Ÿå‡½æ•¸ï¼Œå¯ä»¥ä½¿ç”¨æ–¹æ³• `simplify`ï¼š

```python
x = var('x')
(x^x/x).simplify()
```

ç„¶è€Œï¼Œå°æ–¼æ›´ç´°ç·»çš„ç°¡åŒ–ï¼Œéœ€è¦æ˜ç¢ºæŒ‡å®šæ‰€éœ€çš„ç°¡åŒ–é¡å‹ï¼š

```python
# ç°¡åŒ–å¹³æ–¹æ ¹ã€å°æ•¸æˆ–æŒ‡æ•¸é‹ç®—å¼ï¼Œä½¿ç”¨ canonicalize_radical()ï¼š
f = (e^x-1) / (1+e^(x/2)); f.canonicalize_radical()
```

ä¾‹å¦‚ï¼Œè¦ç°¡åŒ–ä¸‰è§’é‹ç®—å¼ï¼Œå¯ä»¥ä½¿ç”¨æ–¹æ³• `simplify_trig`ï¼š

```python
f = cos(x)^6 + sin(x)^6 + 3 * sin(x)^2 * cos(x)^2
f.simplify_trig()
```

è¦å°‡ä¸‰è§’é‹ç®—å¼ç·šæ€§åŒ–ï¼ˆæˆ–åç·šæ€§åŒ–ï¼‰ï¼Œå¯ä»¥ä½¿ç”¨æ–¹æ³• `reduce_trig`ï¼ˆæˆ– `expand_trig`ï¼‰ï¼š

```python
f = cos(x)^6; f.reduce_trig()
```

```python
f = sin(5 * x); f.expand_trig()
```

åŒ…å«éšä¹˜çš„é‹ç®—å¼ä¹Ÿå¯ä»¥ä½¿ç”¨æ–¹æ³• `simplify_factorial` é€²è¡Œç°¡åŒ–ï¼š

```python
n = var('n'); f = factorial(n+1)/factorial(n)
f.simplify_factorial()
```

ä½¿ç”¨æ–¹æ³• `simplify_rational` å¯ä»¥å˜—è©¦ç°¡åŒ–åˆ†å¼ï¼›è€Œå°æ–¼ç°¡åŒ–å¹³æ–¹æ ¹ã€å°æ•¸æˆ–æŒ‡æ•¸é‹ç®—å¼ï¼Œå»ºè­°ä½¿ç”¨æ–¹æ³• `canonicalize_radical`ï¼š

```python
f = sqrt(abs(x)^2); f.canonicalize_radical()
```

```python
f = log(x*y); f.canonicalize_radical()
```

ä½¿ç”¨æ–¹æ³• `simplify_full` æ™‚ï¼Œæœƒä¾æ¬¡æ‡‰é€²è¡Œ `simplify_factorial`ã€`simplify_rectform`ã€`simplify_trig`ã€`simplify_rational` å’Œ `expand_sum`ã€‚è¦ç¢ºå®šå‡½æ•¸çš„è®ŠåŒ–ï¼ˆå¦‚å°æ•¸ã€æ¼¸è¿‘ç·šã€æ¥µå€¼ã€é›¶é»çš„å®šä½å’Œç¹ªåœ–ï¼‰ï¼Œéƒ½å¯ä»¥è¼•é¬†åœ°é€šéè¨ˆç®—æ©Ÿä»£æ•¸ç³»çµ±ä¾†ç²å¾—ã€‚

### 2.1.5 ä¸€äº›å¯èƒ½çš„éŒ¯èª¤

é€™å°ç¯€æœƒèˆ‰ä¸€äº›ä¾‹å­å±•ç¤ºæ¨™æº–å½¢å¼çš„é‡è¦æ€§ï¼Œå°¤å…¶æ˜¯åœ¨é›¶æ¸¬è©¦ã€‚æŸäº›é‹ç®—å¼å®¶æ—ï¼Œä¾‹å¦‚å¤šé …å¼ï¼Œå…·æœ‰åˆ¤æ–·æ˜¯å¦ç‚ºé›¶çš„æ±ºç­–ç¨‹åºã€‚å› æ­¤ï¼Œå°æ–¼é€™äº›é‹ç®—å¼å®¶æ—ï¼Œç¨‹å¼ç¢¼èƒ½å¤ åˆ¤æ–·çµ¦å®šçš„é‹ç®—å¼æ˜¯å¦ç‚ºé›¶ã€‚åœ¨å¤§å¤šæ•¸æƒ…æ³ä¸‹ï¼Œé€™å€‹åˆ¤æ–·æ˜¯é€šéå°‡é‹ç®—å¼åŒ–ç°¡ç‚ºæ¨™æº–å½¢å¼ä¾†é€²è¡Œçš„ï¼šç•¶ä¸”åƒ…ç•¶é‹ç®—å¼çš„æ¨™æº–å½¢å¼ç‚º 0 æ™‚ï¼Œè©²é‹ç®—å¼ç‚ºé›¶ã€‚

ç„¶è€Œï¼Œä¸¦éæ‰€æœ‰é¡å‹çš„é‹ç®—å¼éƒ½æœ‰æ¨™æº–å½¢å¼ã€‚æ­¤å¤–ï¼Œå°æ–¼æŸäº›é¡å‹çš„é‹ç®—å¼ï¼Œå¯ä»¥è­‰æ˜æ²’æœ‰é€šç”¨æ–¹æ³•èƒ½åœ¨æœ‰é™æ™‚é–“å…§æ±ºå®šä¸€å€‹é‹ç®—å¼æ˜¯å¦ç‚ºé›¶ã€‚é€™é¡é‹ç®—å¼çš„ä¸€å€‹ä¾‹å­åŒ…æ‹¬ç”±æœ‰ç†æ•¸ã€å¸¸æ•¸ ğœ‹ã€log2 å’Œè®Šé‡æ§‹æˆçš„é‹ç®—å¼ï¼Œå†åŠ ä¸ŠåŠ æ¸›ä¹˜æ³•ã€æŒ‡æ•¸å‡½æ•¸å’Œæ­£å¼¦å‡½æ•¸ã€‚åœ¨é€™ç¨®æƒ…æ³ä¸‹ï¼Œåè¦†ä½¿ç”¨ `numerical_approx` æ–¹æ³•ä¸¦æé«˜ç²¾åº¦ï¼Œå¤§å¤šæ•¸æƒ…æ³ä¸‹èƒ½å¤ æ¨æ¸¬ä¸€å€‹é‹ç®—å¼æ˜¯å¦ç‚ºé›¶ï¼›ç„¶è€Œï¼Œå·²ç¶“è­‰æ˜ç„¡æ³•ç·¨å¯«å‡ºä¸€å€‹ç¨‹å¼ç¢¼èƒ½å¤ æ¥å—é€™é¡é‹ç®—å¼ä½œç‚ºè¼¸å…¥ï¼Œä¸¦è¿”å›çœŸï¼ˆå¦‚æœè©²é‹ç®—å¼ç‚ºé›¶ï¼‰æˆ–å‡ï¼ˆå¦‚æœè©²é‹ç®—å¼ä¸ç‚ºé›¶ï¼‰çš„ã€‚

èˆ‰ä¾‹å‡è¨­ ğ‘ ç‚ºä¸€å€‹ç¨å¾®è¤‡é›œçš„é‹ç®—å¼ï¼š

    a = var('a')
    c = (a+1)^2 - (a^2+2*a+1)

åœ¨é€™è£¡ï¼Œæˆ‘å€‘å¸Œæœ›å°è®Šæ•¸ ğ‘¥ è§£æ–¹ç¨‹ ğ‘â‹…ğ‘¥ = 0ï¼š

    eq = c * x == 0 

ç”±æ–¼ c = 0ï¼Œæ•…å…¶è§£ç‚ºä»»æ„ x å€¼ã€‚

åœ¨æ±‚è§£é€™å€‹æ–¹ç¨‹å¼ä¹‹å‰ï¼Œå¯èƒ½æœƒå…ˆå°‡æ–¹ç¨‹å¼å…©é‚ŠåŒæ™‚é™¤ä»¥ ğ‘ï¼š

```python
a, x = var('a, x')
c = (a+1)^2 - (a^2+2*a+1)
eq = c * x == 0 
eq2 = eq / c; eq2
```

```python
solve(eq2, x)
```

ç¾åœ¨æˆ‘å€‘ä¾†è™•ç†æ–¹ç¨‹å¼åŠå…¶è§£æ³•ï¼›ä¸»è¦ä½¿ç”¨çš„æ–¹æ³•ç¸½çµæ–¼è¡¨ 2.3 ä¸­ã€‚

### 2.2.1 é¡¯å¼æ±‚è§£
è®“æˆ‘å€‘è€ƒæ…®ä»¥ä¸‹é€™å€‹æ–¹ç¨‹ï¼Œå…¶ä¸­æœªçŸ¥æ•¸ç‚º $z$ï¼Œåƒæ•¸ç‚º $\phi$ï¼š
$$z^2-\frac{2}{\cos\phi}z+\frac{5}{\cos^2\phi}-4=0,\quad\text{with}\quad\phi\in\left] -\frac{\pi}{2},\frac{\pi}{2} \right[$$
åœ¨ Sage ä¸­ï¼Œå¯ä»¥é€™æ¨£å¯«ï¼š

```python
# è¨­å®š LaTeX é¡¯ç¤ºæ•¸å­¸ç¬¦è™Ÿ
%display latex

z, phi = var('z, phi')
eq = z**2 - 2/cos(phi)*z + 5/cos(phi)**2 - 4 == 0; eq
```

æˆ‘å€‘å¯ä»¥ä½¿ç”¨æ–¹æ³• `lhs`ï¼ˆæˆ– `rhs`ï¼‰ä¾†æå–æ–¹ç¨‹çš„å·¦é‚Šï¼ˆæˆ–å³é‚Šï¼‰ï¼š

```python
eq.lhs()
```

```python
eq.rhs()
```

æ¥è‘—ï¼Œä½¿ç”¨æ–¹æ³• `solve` ä¾†å° $z$ é€²è¡Œæ±‚è§£ï¼š

```python
solve(eq, z)
```
